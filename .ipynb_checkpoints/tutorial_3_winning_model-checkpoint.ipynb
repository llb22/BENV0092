```python
import numpy as np
import pandas as pd
from sklearn.ensemble import HistGradientBoostingRegressor
from sklearn.metrics import mean_absolute_percentage_error
import warnings

LOAD_PATH = "load_data.csv"
TEMP_PATH = "temperature_data.csv"
OUT_PATH  = "submission_load_data.csv"

ZONES = [f"Z{i}" for i in range(1, 22)]

#US public holidays (2005-2007)
US_HOLIDAYS = set(pd.to_datetime([
    # 2005
    "2005-01-01", "2005-01-17", "2005-02-21", "2005-05-30",
    "2005-07-04", "2005-09-05", "2005-10-10", "2005-11-11",
    "2005-11-24", "2005-12-25", "2005-12-26",
    # 2006
    "2006-01-01", "2006-01-02", "2006-01-16", "2006-02-20",
    "2006-05-29", "2006-07-04", "2006-09-04", "2006-10-09",
    "2006-11-10", "2006-11-23", "2006-12-25",
    # 2007
    "2007-01-01", "2007-01-15", "2007-02-19", "2007-05-28",
    "2007-07-04", "2007-09-03", "2007-10-08", "2007-11-12",
    "2007-11-22", "2007-12-25",
]))


def load_data():
    load_df = pd.read_csv(LOAD_PATH, index_col=0, parse_dates=True)
    temp_df = pd.read_csv(TEMP_PATH, index_col=0, parse_dates=True)

    temp_cols = {c: f"temp_{c.strip().replace('station_', 's')}" for c in temp_df.columns}
    temp_df.rename(columns=temp_cols, inplace=True)

    # Merge on datetime index
    df = load_df.join(temp_df, how="left")
    df.index.name = "datetime"
    return df


def engineer_features(df):
    """Build feature matrix from the datetime index and temperature columns."""
    dt = df.index
    temp_cols = [c for c in df.columns if c.startswith("temp_")]

    #Calendar features
    df["hour"]        = dt.hour
    df["dayofweek"]   = dt.dayofweek
    df["month"]       = dt.month
    df["dayofyear"]   = dt.dayofyear
    df["weekofyear"]  = dt.isocalendar().week.astype(int).values
    df["is_weekend"]  = (dt.dayofweek >= 5).astype(int)
    df["is_holiday"]  = dt.normalize().isin(US_HOLIDAYS).astype(int)

    #Cyclical encodings
    df["hour_sin"]   = np.sin(2 * np.pi * df["hour"] / 24)
    df["hour_cos"]   = np.cos(2 * np.pi * df["hour"] / 24)
    df["dow_sin"]    = np.sin(2 * np.pi * df["dayofweek"] / 7)
    df["dow_cos"]    = np.cos(2 * np.pi * df["dayofweek"] / 7)
    df["month_sin"]  = np.sin(2 * np.pi * df["month"] / 12)
    df["month_cos"]  = np.cos(2 * np.pi * df["month"] / 12)
    df["doy_sin"]    = np.sin(2 * np.pi * df["dayofyear"] / 365.25)
    df["doy_cos"]    = np.cos(2 * np.pi * df["dayofyear"] / 365.25)

    #Temperature aggregates
    df["temp_mean"] = df[temp_cols].mean(axis=1)
    df["temp_min"]  = df[temp_cols].min(axis=1)
    df["temp_max"]  = df[temp_cols].max(axis=1)
    df["temp_std"]  = df[temp_cols].std(axis=1)

    #Non-linear temperature
    df["temp_mean_sq"] = df["temp_mean"] ** 2
    df["temp_mean_cb"] = df["temp_mean"] ** 3
    # Heating / Cooling degree-hours (base 65°F)
    df["HDD"] = np.maximum(65.0 - df["temp_mean"], 0)
    df["CDD"] = np.maximum(df["temp_mean"] - 65.0, 0)
    df["HDD_sq"] = df["HDD"] ** 2
    df["CDD_sq"] = df["CDD"] ** 2

    #Temperature lags (24h, 48h, 72h)
    for lag in [24, 48, 72]:
        df[f"temp_mean_lag{lag}"] = df["temp_mean"].shift(lag)

    #Temperature rolling averages
    df["temp_mean_roll24"]  = df["temp_mean"].rolling(24, min_periods=1).mean()
    df["temp_mean_roll168"] = df["temp_mean"].rolling(168, min_periods=1).mean()  # 7 days

    #Interaction terms
    df["hour_x_temp"]    = df["hour"] * df["temp_mean"]
    df["weekend_x_hour"] = df["is_weekend"] * df["hour"]
    df["month_x_temp"]   = df["month"] * df["temp_mean"]
    df["holiday_x_hour"] = df["is_holiday"] * df["hour"]

    #Trend
    df["trend"] = np.arange(len(df))

    return df


def get_feature_cols(df):
    """Return the list of feature column names"""
    return [c for c in df.columns if c not in ZONES]


def train_and_predict(df, zone, feature_cols, verbose=True):
    # Split: 2005 = train, 2006 = validation, 2007 = forecast target
    mask_2005 = df.index.year == 2005
    mask_2006 = df.index.year == 2006
    mask_train = mask_2005 | mask_2006  #full training data
    mask_2007 = df.index.year == 2007

    #Validation pass (train on 2005, validate on 2006)
    X_train_val = df.loc[mask_2005, feature_cols].values
    y_train_val = df.loc[mask_2005, zone].values
    X_val       = df.loc[mask_2006, feature_cols].values
    y_val       = df.loc[mask_2006, zone].values

    model_val = HistGradientBoostingRegressor(
        loss="poisson",
        learning_rate=0.05,
        max_iter=1500,
        max_depth=8,
        max_leaf_nodes=63,
        min_samples_leaf=20,
        l2_regularization=0.1,
        early_stopping=True,
        validation_fraction=None,
        n_iter_no_change=50,
        random_state=42,
    )
    model_val.fit(X_train_val, y_train_val)
    val_pred = model_val.predict(X_val)
    val_mape = mean_absolute_percentage_error(y_val, val_pred) * 100

    if verbose:
        print(f"  {zone}: Validation MAPE = {val_mape:.2f}%")

    #Final model (train on 2005+2006, predict 2007)
    X_full = df.loc[mask_train, feature_cols].values
    y_full = df.loc[mask_train, zone].values

    model_final = HistGradientBoostingRegressor(
        loss="poisson",
        learning_rate=0.05,
        max_iter=model_val.n_iter_,  #same number of iterations
        max_depth=8,
        max_leaf_nodes=63,
        min_samples_leaf=20,
        l2_regularization=0.1,
        early_stopping=False,
        random_state=42,
    )
    model_final.fit(X_full, y_full)

    X_2007 = df.loc[mask_2007, feature_cols].values
    pred_2007 = model_final.predict(X_2007)

    return val_mape, pred_2007


def main():
    # 1. Load & merge data
    df = load_data()
    print(f"      Loaded {len(df)} rows × {len(df.columns)} columns")

    # 2. Feature engineering
    df = engineer_features(df)
    feature_cols = get_feature_cols(df)
    print(f"      {len(feature_cols)} features")

    # 3. Train per-zone models
    mapes = {}
    preds_2007 = {}

    for zone in ZONES:
        mape, pred = train_and_predict(df, zone, feature_cols)
        mapes[zone] = mape
        preds_2007[zone] = pred

    #4 - Summary stats
    mean_mape = np.mean(list(mapes.values()))
    print(f"  Mean validation MAPE: {mean_mape:.2f}%")
    print(f"  Z21 validation MAPE:  {mapes['Z21']:.2f}%")

    #5 Make Submission CSV to send
    load_df = pd.read_csv(LOAD_PATH, index_col=0, parse_dates=True)
    mask_2007 = load_df.index.year == 2007

    for zone in ZONES:
        load_df.loc[mask_2007, zone] = preds_2007[zone]

    #Sanity check
    num_nan = load_df.isna().sum().sum()
    print(f"NaN cells: {num_nan}")

    load_df.to_csv(OUT_PATH)
    print(f"Saved → {OUT_PATH}")

main()
```
